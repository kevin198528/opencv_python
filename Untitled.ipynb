```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='A')
plt.plot(x, x**2, label='B')
plt.plot(x, x**3, label='C')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()

plt.show()
```

```python
def my_plotter(ax, data_x, data_y, **param_dict):
    ax.plot(data_x, data_y, **param_dict)
    
data1, data2, data3, data4 = np.random.randn(4, 100)

fig, (ax1, ax2) = plt.subplots(2, 1)

my_plotter(ax1, data1, data2, marker='o')
my_plotter(ax2, data3, data4, marker='o')
```
